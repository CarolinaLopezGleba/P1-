## Parte 1 -  Nombres populares de todos los años agrupados 

En lugar de tratar los nombres de niños y niñas por separado, los agruparemos por completo. En algunos años, un nombre aparece más de una vez en el código HTML, pero solo utilizaremos un número por nombre (el primero que aparezca).


```python
import re

# Lista para recorrer todos los archivos
nombres_archivos = ['baby1990.html', 'baby1992.html', 'baby1994.html','baby1996.html','baby1998.html','baby2000.html','baby2002.html','baby2004.html','baby2006.html', 'baby2008.html']

# Definición de patrones de busqueda de nombres de niñas y niños de los diferentes archivos 
patron_ninias = '<tr align="right"><td>\d+</td><td>[A-Za-z]+</td><td>([A-Za-z]+)</td>'
patron_ninios = '<tr align="right"><td>\d+</td><td>([A-Za-z]+)</td><td>[A-Za-z]+</td>'

# Listas para guardar patrones encontrados
nombres_ninias_completo = []
nombres_ninios_completo = []

# Recorre cada linea de cada archivo en la lista
for line in nombres_archivos:
    archivo = open(line)
    texto = archivo.read()
    
    # Realiza la búsqueda del patrón en el contenido del archivo definido para nombres de niñas
    nombres_ninias = re.findall(patron_ninias, texto)
    
    # Realiza la búsqueda del patrón en el contenido del archivo definido para nombres de niños
    nombres_ninios = re.findall(patron_ninios, texto)

    # Si se encontraron coincidencias para el patrón niñas, imprime los nombres encontrados
    if nombres_ninias:
        print(f"Archivo: {line}")
        print("Nombres de niña:")
        
        for nombre in nombres_ninias:

            print(nombre)
            nombres_ninias_completo.append(nombre)

    # Si se encontraron coincidencias para el patrón niños, imprime los nombres encontradoss
    if nombres_ninios:
        print(f"Archivo: {line}")
        print("Nombres de niño:")
        for nombre in nombres_ninios:

            print(nombre)
            nombres_ninios_completo.append(nombre)

# Filtra los nombres que aparecen más de una vez para cada patrón definido
ninias_filtro=[]
ninios_filtro =[]
for nombre in nombres_ninias_completo:

    if nombre not in ninias_filtro:
        ninias_filtro.append(nombre)

for nombre in nombres_ninios_completo:

    if nombre not in ninios_filtro:
        ninios_filtro.append(nombre)


print(ninias_filtro)
print(ninios_filtro)
```

## Parte 2 - Año de cada archivo

Utilizando expresiones regulares debes encontrar el año, extraerlo y mostrarlo.

```python
patron_anio = 'Popularity in (\d{4})'

# Diccionario para almacenar el patrón definido
anio_por_archivo = {}

# Recorrer cada linea de cada archivo en la lista definida de archivos
for line in nombres_archivos:
    
    with open(line) as archivo:
        texto = archivo.read()
        # Encontrar todas las coincidencias con el patrón en el contenido del archivo
        patron_anio_encontrado = re.findall(patron_anio, texto)
        anio_por_archivo[line] = patron_anio_encontrado
        
# Imprimir el diccionario con el patrón definido para año
for archivo, anio_archivo in anio_por_archivo.items():

    print(f"Archivo: {archivo}")
    print(f"Año del archivo: {anio_archivo}")
    print("\n")
```

## Parte 3 - Nombres y su ranking 

Extrae los nombres de niños y de niñas con el correspondiente número que ocupan en el ranking y muéstralos.

```python
# Definición de patrones para busqueda de rankings y nombres de niña y niño 
patron_ranking_ninias = '<tr align="right"><td>(\d+)</td><td>[A-Za-z]+</td><td>([A-Za-z]+)</td>'
patron_ranking_ninios = '<tr align="right"><td>(\d+)</td><td>([A-Za-z]+)</td><td>[A-Za-z]+</td>'

# Recorrer cada linea de cada archivo
for archivo_nombre in nombres_archivos:
    
    archivo = open(archivo_nombre)
    texto = archivo.read()
    
    # Realiza la búsqueda del patrón definido para niñas en cada archivo
    encontrados_ninias = re.findall(patron_ranking_ninias, texto)

    # Realiza la búsqueda del patrón definido para niños en cada archivo 
    encontrados_ninios = re.findall(patron_ranking_ninios, texto)
    
    # Si se encontraron coincidencias para nombres de niñas imprime el nombre del archivo y las coincidencias con el patron definido
    if encontrados_ninias:
        print(f"Archivo: {archivo_nombre}")
        print("Patrón niñas:")
        for patron in encontrados_ninias:
            print(patron)

    # Si se encontraron coincidencias para nombres de niños, imprime el nombre del archivo y las coincidencias con el patron definido
    if encontrados_ninios:
        print(f"Archivo: {archivo_nombre}")
        print("Patrón niños:")
        for patron in encontrados_ninios:
            print(patron)
```

## Parte 4 - Nombres y su ranking - diccionario e impresión

Guarda los nombres (individuales) y su ranking en un diccionario e imprímelos.

```python
# Crea un diccionario para almacenar los resultados
diccionario = {}

# Recorrer cada linea de cada archivo
for archivo_nombre in nombres_archivos:
    
    archivo = open(archivo_nombre)
    texto = archivo.read()
    
    # Realiza la búsqueda del patrón definido para niñas en cada archivo
    coincidencia_ninias = re.findall(patron_ranking_ninias, texto)

    # Realiza la búsqueda del patrón definido para niños en cada archivo 
    coincidencia_ninios = re.findall(patron_ranking_ninios, texto)
    
    # Si se encontraron coincidencias para nombres de niñas, guárdalas en el diccionario
    if coincidencia_ninias:
        diccionario[archivo_nombre] = {
            "Patrón niñas": coincidencia_ninias
        }

    # Si se encontraron coincidencias para nombres de niños, guárdalas en el diccionario
    if coincidencia_ninios:
        if archivo_nombre in diccionario:
            diccionario[archivo_nombre]["Patrón niños"] = coincidencia_ninios
        else:
            diccionario[archivo_nombre] = {
                "Patrón niños": coincidencia_ninios
            }
print(diccionario)
```

## Parte 5 - Nombres ordenados alfabeticamente - lista e impresión

Crea una lista con los nombres ordenados alfabéticamente y muéstrala.

```python
# Lista para almacenar las coincidencias de nombres
lista_nombres_ordenados = []

# Recorrer cada linea de cada archivo 
for archivo_nombre in nombres_archivos:
    archivo = open(archivo_nombre)
    texto = archivo.read()

    # Realiza la búsqueda del patrón en el contenido del archivo para el patrón niños
    ninios_encontrados = re.findall(patron_ninios, texto)

    # Realiza la búsqueda del patrón en el contenido del archivo para el patrón niñas
    ninias_encontradas = re.findall(patron_ninias, texto)

    # Si se encontraron coincidencias para el patrón niños, agregalas a la lista 
    if ninios_encontrados:
        for nombre in ninios_encontrados:
            lista_nombres_ordenados.append(nombre)

    # Si se encontraron coincidencias para el niñas, agregalas a la lista 
    if ninias_encontradas:
        for nombre in ninias_encontradas:
            lista_nombres_ordenados.append( nombre)

# Ordenar la lista alfabéticamente
nombres_ordenados = sorted(lista_nombres_ordenados)

# Imprimir los nombres ordenados
for nombre in nombres_ordenados:
    print(nombre)
```

## Parte 6 - Año, nombre y ranking

Crea una lista con el siguiente formato y muéstrala:

• El primer elemento debe contener un string con el "año",

• Los siguientes elementos deben contener un string con el "nombre y la posición del ranking" ordenados en forma alfabética.

```python
# Lista para almacenar las coincidencias de nombres, rankings y años
coincidencias_nombres_rankings_anios = []

# Recorrer cada linea de cada archivo
for archivo_nombre in nombres_archivos:
    archivo = open(archivo_nombre)
    texto = archivo.read()

    # Realiza la búsqueda del patrón niños en el contenido de los archivos
    ninios_encontrados = re.findall(patron_ranking_ninios, texto)

    # Realiza la búsqueda del patrón niñas en el contenido de los archivos
    ninias_encontradas = re.findall(patron_ranking_ninias, texto)

    # Si se encontraron coincidencias para el patrón 1, agregalas a la lista de coincidencias con el año
    if ninios_encontrados:
        for ranking, nombre in ninios_encontrados:
            coincidencias_nombres_rankings_anios.append((anio_por_archivo[archivo_nombre], (ranking), nombre))

    # Si se encontraron coincidencias para el patrón 2, agregalas a la lista de coincidencias con el año
    if ninias_encontradas:
        for ranking, nombre in ninias_encontradas:
            coincidencias_nombres_rankings_anios.append((anio_por_archivo[archivo_nombre], (ranking), nombre))

# Definir una función personalizada para el ordenamiento
def ordenar_coincidencias(x):
    return (x[2], x[0])

# Ordenar la lista de patrones definidos alfabéticamente por nombre y luego por año
coincidencias_nombres_rankings_anios.sort(key=ordenar_coincidencias)

# Imprimir la lista ordenada con el primer elemento año como string, nombre y ranking en string. 
for anio, ranking, nombre in coincidencias_nombres_rankings_anios:
    print(f"Año Nombre y Ranking:  {anio} {nombre, ranking}")
```
